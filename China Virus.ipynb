```python
#reduce data to only worst hit countries
"""countries_list = ['Belgium',
                 'UK',
                 'Peru',
                 'Spain',
                 'Italy',
                 'Sweden',
                 'Chile',
                 'US',
                 'Brazil',
                 'France',
                 'Mexico',
                 'US Without NY NJ']
"""
countries_list = ['US Only NY',
                  'US Only NJ',
                  'US Only FL',
                  'US Only TX']
country_df = country_df[country_df['Country_Region'].isin(countries_list)]
```

```python
countries_list = sorted(list(country_df['Country_Region'].unique()))

#colors_list = ["gold", "blue", "green", "yellow", "black", "grey", "darkgreen", "pink", "brown", "slateblue", "grey1", "orange"]
#colors_list = ["grey", "grey", "grey", "grey", "grey", "grey", "grey", "grey", "grey", "grey", "blue", "blue"]
#colors_dict = dict(zip(countries_list,colors_list))
colors_dict = {'US Only NY':'purple',
              'US Only NJ':'blue',
              'US Only FL':'green',
              'US Only TX':'orange'}

rename_dict = {'US Only NY':'New York',
              'US Only NJ':'New Jersey',
              'US Only FL':'Florida',
              'US Only TX':'Texas'}


data = []

for mycountry in countries_list:
    data.append(go.Scatter(x=country_df[country_df['Country_Region']==mycountry]['Date'],
                               y=country_df[country_df['Country_Region']==mycountry]['Deaths_Per_Million'],
                               name = rename_dict[mycountry],
                               line = {'color':colors_dict[mycountry]}
                          )
                )


layout = go.Layout(
                showlegend = True,
                legend= dict(x=0.03,y=0.97),
                title='Coronavirus Deaths Per Million',
                title_x= 0.5,
                yaxis={'title':'Deaths_Per_Million',
                       'zeroline':True,'zerolinecolor':'#000000','zerolinewidth':2.0},
                template='plotly_white')

fig = go.Figure(data=data,layout=layout)

#fig.update_layout()

iplot(fig)
```
